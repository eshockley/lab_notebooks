```python
def draw_lhs_from_mcmc_estimated_pdf(nsamples, mcmc_df):
    final_samples = np.zeros((nsamples, len(mcmc_df.columns)))
    quantiles = np.linspace(0, 1, num=nsamples+1)
    for i, qt in enumerate(quantiles[:len(quantiles)-1]):
        next_qt = quantiles[i+1]
        subsamples = mcmc_df[(mcmc_df[mcmc_df.columns[0]] >= mcmc_df.quantile(qt)[mcmc_df.columns[0]]) & (mcmc_df[mcmc_df.columns[0]] <= mcmc_df.quantile(next_qt)[mcmc_df.columns[0]])]
        subsamples_mean = subsamples.mean().values
        #cov = np.cov(subsamples.values, rowvar=0)
        final_samples[i] = subsamples_mean
    
    return final_samples
```

```python
lhs_samples = draw_lhs_from_mcmc_estimated_pdf(12000, df)
```

```python
def create_more_samples_beyond_posterior(params_to_expand, mcmc_df, n10quants_to_expand, points_per_10quant):
    npoints_to_add = len(params_to_expand)*2*points_per_10quant
    added_points = np.zeros((npoints_to_add, len(mcmc_df.columns)))
    bottom_quant = n10quants_to_expand*.1
    top_quant = 1-(n10quants_to_expand*.1)
    startidx = 0
    endidx = startidx + points_per_10quant
    for param in params_to_expand:
        min_point = mcmc_df.quantile(0)
        max_point = mcmc_df.quantile(1)
        bottom_quant_pts = mcmc_df[(mcmc_df[param] >= mcmc_df.quantile(0)[param]) & (mcmc_df[param] <= mcmc_df.quantile(bottom_quant)[param])]
        top_quant_pts = mcmc_df[(mcmc_df[param] >= mcmc_df.quantile(top_quant)[param]) & (mcmc_df[param] <= mcmc_df.quantile(1)[param])]
        mean = bottom_quant_pts.mean().values
        cov = np.cov(bottom_quant_pts.values, rowvar=0)
        pts_to_add = np.random.multivariate_normal(mean, cov)
        added_points[startidx:endidx, :] = pts_to_add
        startidx = endidx
        endidx += points_per_10quant
        mean = top_quant_pts.mean().values
        cov = np.cov(top_quant_pts.values, rowvar=0)
        pts_to_add = np.random.multivariate_normal(mean, cov)
        added_points[startidx:endidx, :] = pts_to_add
        startidx = endidx
        endidx += points_per_10quant
    
    df = pd.DataFrame(added_points, columns=mcmc_df.columns)
    
    nlhs_points = len(added_points)/100
    
    lhs_points = draw_lhs_from_mcmc_estimated_pdf(nlhs_points, df)
    
    return lhs_points
```

```python
params_to_expand = ['kcat_AA2', 'kcat_AA3', 'KD_AG_allo2', 'KD_AA_allo2', 'KD_AA_allo3', 'kcat_AG3']
points_to_add = create_more_samples_beyond_posterior(params_to_expand, lhs_df.dropna(), 2, 10000)
```

```python
lhs_points_to_add = create_more_samples_beyond_posterior(params_to_expand, lhs_df.dropna(), 2, 100000)
```
