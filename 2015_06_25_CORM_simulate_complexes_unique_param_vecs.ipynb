```python
import pickle
from helper_fxns import create_trace_matrix, find_unique_param_vecs, sub_parameters, plot_histograms
from corm import model
from pysb.integrate import Solver
import numpy as np
%matplotlib inline
```

```python
old_rates_data = pickle.load(open('2015_06_22_COX2_rerun_with_old_rates_oldver.p'))
```

```python
old_rates_arr = create_trace_matrix(old_rates_data, burnin=150000, thin=1)
```

```python
old_rates_uniques = find_unique_param_vecs(old_rates_arr)
len(old_rates_uniques)
unique_vecs, inv_idx = np.unique(old_rates_arr.view([('', old_rates_arr.dtype)]*old_rates_arr.shape[1]), return_inverse=True)
```

```python
simulation_data = np.zeros((len(old_rates_uniques), 8))
```

```python
model.observables
```

```python
t = np.linspace(0, 10, num=100)
solver = Solver(model, t)
model.parameters['AG_0'].value = 2
model.parameters['AA_0'].value = 2
observables = ['obsAAallo', 'obsAAcat', 'obsAAboth', 'obsAGallo', 'obsAGcat', 'obsAGboth', 'obsAAcatAGallo', 'obsAGcatAAallo']
```

```python
for nvec, unique_parameter_set in enumerate(old_rates_uniques):
    param_dict = {}
    for i, param in enumerate(old_rates_data.keys()):
        param_dict[param] = unique_parameter_set[i]
    sub_parameters(model, param_dict, verbose=False)
    solver.run()
    for n, observable in enumerate(observables):
        simulation_data[nvec][n] = solver.yobs[observable][-1]
    if nvec % 1000 == 0:
        print 'Simulated ',nvec,' parameter vectors.'
```

```python
log_sim_data = np.log10(simulation_data)
simulation_data
```

```python
full_sim_data = log_sim_data[inv_idx]
```

```python
plot_histograms(observables, full_sim_data)
```

```python
original_data = pickle.load(open('2015_01_27_Dream_COX2_parameter_fits.p'))
```

```python
original_data_arr = create_trace_matrix(original_data, burnin=100000, thin=10)
```

```python
original_data_uniques = find_unique_param_vecs(original_data_arr)
len(original_data_uniques)
```

```python
unique_vecs, inv_idx = np.unique(original_data_arr.view([('', original_data_arr.dtype)]*original_data_arr.shape[1]), return_inverse=True)
```

```python
original_sim_data = np.zeros((len(original_data_uniques), 9))
```

```python
full_observables = ['obsAAallo', 'obsAAcat', 'obsAAboth', 'obsAGallo', 'obsAGcat', 'obsAGboth', 'obsAAcatAGallo', 'obsAGcatAAallo', 'obsdifflig']
exp_cond_AA = [.5, 1, 2, 4, 8, 16]
exp_cond_AG = [.5, 1, 2, 4, 8, 16]

for AA_init in exp_cond_AA:
        for AG_init in exp_cond_AA:
            print 'Simulating AA = '+str(AA_init)+' AG = '+str(AG_init)
            model.parameters['AA_0'].value = AA_init
            model.parameters['AG_0'].value = AG_init
            for nvec, unique_parameter_set in enumerate(original_data_uniques):
                param_dict = {}
                for i, param in enumerate(original_data.keys()):
                    param_dict[param] = unique_parameter_set[i]
                sub_parameters(model, param_dict, verbose=False)
                solver.run()
                for n, observable in enumerate(observables):
                    original_sim_data[nvec][n] = solver.yobs[observable][-1]
                original_sim_data[nvec][8] = original_sim_data[nvec][6] + original_sim_data[nvec][7]
                if nvec % 1000 == 0:
                    print 'Simulated ',nvec,' parameter vectors.'
            log_original_sim_data = np.log10(original_sim_data)
            full_original_sim_data = log_original_sim_data[inv_idx]
            print 'Histogram for AA = ',str(AA_init),' AG = ',str(AG_init),' : '
            plot_histograms(full_observables, full_original_sim_data)
```

```python
model.parameters['AA_0'].value = 2
model.parameters['AG_0'].value = 2
for nvec, unique_parameter_set in enumerate(original_data_uniques):
    param_dict = {}
    for i, param in enumerate(original_data.keys()):
        param_dict[param] = unique_parameter_set[i]
    sub_parameters(model, param_dict, verbose=False)
    solver.run()
    for n, observable in enumerate(observables):
        original_sim_data[nvec][n] = solver.yobs[observable][-1]
    if nvec % 1000 == 0:
        print 'Simulated ',nvec,' parameter vectors.'
    log_original_sim_data = np.log10(original_sim_data)
    full_original_sim_data = original_sim_data[inv_idx]
```
