```python
df = pd.read_csv('2016_02_24_HeLa_raw_results.csv', usecols=[1, 0, 2, 3, 9])
hier_df = df.set_index(['Protein Name', 'Peptide Sequence', 'Replicate Name'])
hier_df_nodup = hier_df.drop_duplicates()
```

```python
final_data_df = pd.DataFrame(index=hier_df_nodup.index, columns=['Light Area', 'Heavy Area', 'Heavy:Light Ratio'])
final_data_df.index[0]
```

```python
for idx in hier_df_nodup.index:
    replicate_arr = hier_df_nodup.ix[idx]
    light_mz = replicate_arr.ix[idx]['Precursor Mz'].min()
    heavy_mz = replicate_arr.ix[idx]['Precursor Mz'].max()
    light_arr = replicate_arr.loc[replicate_arr['Precursor Mz'] == light_mz]
    heavy_arr = replicate_arr.loc[replicate_arr['Precursor Mz'] == heavy_mz]
    light_area = light_arr.get_value(light_arr.index[0], 'Total Area')
    heavy_area = heavy_arr.get_value(heavy_arr.index[0], 'Total Area')
    final_data_df.set_value(idx, 'Light Area', light_area)
    final_data_df.set_value(idx, 'Heavy Area', heavy_area)
    final_data_df.set_value(idx, 'Heavy:Light Ratio', heavy_area/light_area)
```

```python
final_data_df_nodups = final_data_df.drop_duplicates()
```

```python
calibration_curve_ratios = pd.read_csv('Light Ratios.csv', usecols=range(0,10))
calibration_curve_ratios
```

```python
loq_lod_results = pd.read_csv('LOQ_and_LOD.csv')
loq_lod_results
```

```python
calibration_curve_summary = calibration_curve_ratios
calibration_curve_summary.insert(10, 'LOQ', loq_lod_results['LOQ'])
```

```python
calibration_curve_summary.insert(11, 'LOD', loq_lod_results['LOD'])
```

```python
replacement_mapping_dict = {'LOQ':{0:'No linear response', 4e-18:'4amol', 4e-17:'40amol', 4e-16:'400amol', 2e-15:'2fmol', 4e-15:'4fmol', 1e-14:'10fmol', 2e-14:'20fmol', 4e-14:'40fmol'}, 'LOD':{0:'No linear response', 4e-18:'4amol', 4e-17:'40amol', 4e-16:'400amol', 2e-15:'2fmol', 4e-15:'4fmol', 1e-14:'10fmol', 2e-14:'20fmol', 4e-14:'40fmol'}}
```

```python
calibration_curve_summary.replace(to_replace=replacement_mapping_dict, inplace=True)
```

```python
calibration_curve_summary.insert(12, 'Slope', None)
calibration_curve_summary.insert(13, 'Intercept', None)
calibration_curve_summary.insert(14, 'Rsquared', None)
```

```python
from scipy.stats import linregress
import numpy as np
```

```python
log_spike_vals = np.log10([4e-18, 4e-17, 4e-16, 2e-15, 4e-15, 1e-14, 2e-14, 4e-14])
spike_labels = ['4amol', '40amol', '400amol', '2fmol', '4fmol', '10fmol', '20fmol', '40fmol']
ncol_spike_dict = {'4amol':2, '40amol':3, '400amol':4, '2fmol':5, '4fmol':6, '10fmol':7,'20fmol':8, '40fmol':9}

for row in range(len(calibration_curve_summary)):
    y_vals = calibration_curve_summary.lookup(row_labels=[row]*len(spike_labels), col_labels=calibration_curve_summary.columns[ncol_spike_dict['4amol']:ncol_spike_dict['40fmol']+1])
    x_vals = log_spike_vals
    slope, intercept, r_value, p_value, std_err = linregress(x_vals, y_vals)
    print 'R squared with all points: ',r_value**2
    col_label = '4amol'
    n_line_pts = len(log_spike_vals) - 1
    ncol = ncol_spike_dict['40amol']
    while np.nan_to_num(r_value**2) < .85 or r_value < 0:
        y_vals = calibration_curve_summary.lookup(row_labels=[row]*n_line_pts, col_labels=calibration_curve_summary.columns[ncol:ncol_spike_dict['40fmol']+1])
        x_vals = log_spike_vals[-n_line_pts::]
        slope, intercept, r_value, p_value, std_err = linregress(x_vals, y_vals)
        print 'R squared: ',r_value**2,' with npoints: ',n_line_pts
        col_label = spike_labels[-n_line_pts]
        n_line_pts -= 1
        print 'n_line_pts: ',n_line_pts
        ncol += 1
        if n_line_pts == 2:
            break
    
    if r_value**2 > .85 and r_value > 0:
        calibration_curve_summary.set_value(row, 'LOQ', col_label)
        calibration_curve_summary.set_value(row, 'LOD', '4amol')
        calibration_curve_summary.set_value(row, 'Slope', slope)
        calibration_curve_summary.set_value(row, 'Intercept', intercept)
        calibration_curve_summary.set_value(row, 'Rsquared', r_value**2)
        print 'For peptide: ',calibration_curve_summary.loc[row]['peptide_seq'],' n points = ',n_line_pts,' R2 = ',calibration_curve_summary.loc[row]['Rsquared']
    else:
        calibration_curve_summary.set_value(row, 'LOQ', 'No linear range')
        calibration_curve_summary.set_value(row, 'LOD', 'No linear range')
        calibration_curve_summary.set_value(row, 'Slope', None)
        calibration_curve_summary.set_value(row, 'Intercept', None)
        calibration_curve_summary.set_value(row, 'Rsquared', r_value**2)
        print 'Peptide: ',calibration_curve_summary.loc[row]['peptide_seq'],' had no linear range.'
```

```python
calibration_curve_summary.set_index(['peptide_seq'], inplace=True)
```

```python
def calc_absolute_conc(measured_ratio, calibration_curve_slope, calibration_curve_intercept):
    return (measured_ratio - calibration_curve_intercept)/calibration_curve_slope
```

```python
final_data_df_nodups.insert(3, 'Calculated Moles', None)
```

```python
final_data_df_nodups.insert(4, 'Failure Reason', 'No failure.')
```

```python
spike_dict = {label:val for label,val in zip(spike_labels, log_spike_vals)}
for idx in final_data_df_nodups.index:
    measured_ratio = final_data_df_nodups.get_value(idx, 'Heavy:Light Ratio')
    try:
        calib_curve_idx = idx[1]
        slope = calibration_curve_summary.get_value(calib_curve_idx,'Slope')
        
    except KeyError:
        for nclip in range(1, len(idx[1])):
            calib_curve_idx = idx[1][0:-nclip]
            try:
                slope = calibration_curve_summary.get_value(calib_curve_idx, 'Slope')
                break
            except KeyError:
                slope == None
        
    if slope == None:
        final_data_df_nodups.set_value(idx, 'Calculated Moles', np.nan)
        final_data_df_nodups.set_value(idx, 'Failure Reason', 'No linear range.')
        print 'idx: ',idx,' Set to no linear range.'
        
    else:
        intercept = calibration_curve_summary.get_value(calib_curve_idx, 'Intercept')
        calculated_moles = 10**calc_absolute_conc(measured_ratio, slope, intercept)
        loq = calibration_curve_summary.ix[calib_curve_idx]['LOQ']
        loq_val = spike_dict[loq]
        print 'calculated moles: ',calculated_moles
        if calculated_moles < loq_val:
            final_data_df_nodups.set_value(idx, 'Calculated Moles', 'Below quantitation limit.')
            final_data_df_nodups.set_value(idx, 'Failure Reason', 'Below quantitation limit.')
            print 'Calculated moles for idx: ',idx,' set to below quantitation limit.'
        else:
            final_data_df_nodups.set_value(idx, 'Calculated Moles', calculated_moles)
            final_data_df_nodups.set_value(idx, 'Failure Reason', 'No failure.')
        
```

```python
final_data_df_nodups.to_csv('2016_03_02_HeLa_timecourse_quantitation.csv')
```
