```python
samples = [np.load('ndimgauss_mtdreamzs_3chain_sampled_params_chain_'+str(chain)+'.npy') for chain in range(3)]
```

```python
samples = np.concatenate((samples[0][75000:, :], samples[1][75000:,:], samples[2][75000:,:]))
```

```python
samples[:,0].shape
```

```python
d = 200
A = .5 * np.identity(d) + .5 * np.ones((d,d))
C = np.zeros((d,d))
for i in range(d):
    for j in range(d):
        C[i][j] = A[i][j] * np.sqrt((i+1)*(j+1))
mu = np.zeros(d)

perfect_samples = np.random.multivariate_normal(mean=mu, cov=C, size=225000)
```

```python
colors = sns.color_palette(n_colors=200)
for dim in range(200):
    fig = plt.figure()
    ax = sns.distplot(samples[:,dim], color=colors[dim])
    sns.distplot(perfect_samples[:,dim], color='k', hist=False)
    plt.xlim([-50, 50])
    plt.ylim(0, .5)
    fig.savefig('ndimgauss_model_appnoteplot_dim'+str(dim))
```
