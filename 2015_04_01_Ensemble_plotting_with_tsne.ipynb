```python
import numpy as np
from fast_tsne import fast_tsne
from PCA import PCA
import matplotlib
%matplotlib inline
```

# Case 1: One Basin

## Test 1

10d parameter vectors.  First and simplest test: all means are set equal to 0 and pairwise covariances are .5.  Standard deviations are equal to the dimension number (i.e. 1-10):

```python
simple_cov_matrix = np.full((10, 10), .5)
np.fill_diagonal(simple_cov_matrix, np.linspace(1, 10, num=10))
```

```python
simple_ensemble = np.random.multivariate_normal(np.array([0]*10), simple_cov_matrix, 5000)
```

```python
tsne_single_ensemble_simple = fast_tsne(simple_ensemble)
pca_single_ensemble_simple = PCA(simple_ensemble)
```

```python
pca_single_ensemble_simple.shape
```

```python
matplotlib.pyplot.scatter(tsne_single_ensemble_simple[:,0], tsne_single_ensemble_simple[:,1])
```

Are the occasional 'islands' off the main blob due to not enough random samples (i.e. they should be connected to the main island but by random chance I didn't sample any parameter vectors that would fall in the connecting area)?  Checking:

```python
larger_ensemble = np.random.multivariate_normal(np.array([0]*10), simple_cov_matrix, 50000)
```

```python
tsne_single_ensemble_simple_larger = fast_tsne(larger_ensemble)
```

```python
matplotlib.pyplot.scatter(tsne_single_ensemble_simple_larger[:,0], tsne_single_ensemble_simple_larger[:,1])
```

Now the islands are gone, so that was just a side effect of a small number of random samples.

Checking PCA output:

```python
matplotlib.pyplot.scatter(pca_single_ensemble_simple[0,:], pca_single_ensemble_simple[1,:])
```

## Test 3

What if I create an ensemble where the parameters are drawn from distributions centered around the same location and with the same sd, but where there are two ways in which the parameters can covary and be consistent with the experimental data?

```python
alt_cov_matrix = np.random.random_sample(size=(10,10))
np.fill_diagonal(alt_cov_matrix, np.linspace(1, 10, num=10))
alt_ensemble = np.random.multivariate_normal(np.array([0]*10), alt_cov_matrix, 5000)
```

```python
merged_ensemble = np.concatenate((simple_ensemble2, alt_ensemble), axis=0)
```

```python
matplotlib.pyplot.scatter(tsne_merged_ensemble[:,0], tsne_merged_ensemble[:,1])
matplotlib.pyplot.figure()
matplotlib.pyplot.scatter(pca_merged_ensemble[0,:], pca_merged_ensemble[1,:])
```

I've already plotted the 'simple_ensemble2' half of the merged ensemble above.  What does the 'alt_ensemble' half look like by itself?

```python
tsne_alt_ensemble = fast_tsne(alt_ensemble)
pca_alt_ensemble = PCA(alt_ensemble)
```

```python
matplotlib.pyplot.scatter(tsne_alt_ensemble[:,0], tsne_alt_ensemble[:,1])
matplotlib.pyplot.figure()
matplotlib.pyplot.scatter(pca_alt_ensemble[0,:], pca_alt_ensemble[1,:])
```

```python
avg_cov_matrix = (simple_cov_matrix2 + alt_cov_matrix)/2
```

```python
avg_cov_ensemble = np.random.multivariate_normal(np.array([0]*10), avg_cov_matrix, 10000)
```

```python
tsne_avg_ensemble = fast_tsne(avg_cov_ensemble)
pca_avg_ensemble = PCA(avg_cov_ensemble)
```
