```python
from projection_fxns import calc_normalized_shape_idx
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
sphere_means = np.array([10, -1, 0])
sphere_covariance = np.array([[3, 0, 0], [0, 3, 0], [0, 0, 3]])
sphere_samples = np.random.multivariate_normal(mean=sphere_means, cov=sphere_covariance, size=500000)
ranges = np.ptp(sphere_samples, axis=0)
norm_sphere_samples = sphere_samples/ranges
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(sphere_samples[:,0], sphere_samples[:,1], sphere_samples[:,2])
ax2 = fig.add_subplot(111, projection='3d')
ax2.scatter(norm_sphere_samples[:,0], norm_sphere_samples[:,1], norm_sphere_samples[:,2])
```

```python
calc_normalized_shape_idx(sphere_samples, approx=True, verbose=True)
```

```python
calc_normalized_shape_idx(norm_sphere_samples, verbose=True)
```
